## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df

```

```python
# Create a list of at least 10 column names to use as X data
sel_x_columns =  ['Age', 'Education','HourlyRate','DistanceFromHome',
                               'OverTime','StockOptionLevel','WorkLifeBalance' ,'YearsAtCompany',
                               'YearsSinceLastPromotion','NumCompaniesWorked']




# Create X_df using your selected columns
X_df = attrition_df[sel_x_columns]


# Show the data types for X_df
X_df.info()

```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X_df, y_df, test_size=0.2, random_state=42)

```

```python
from sklearn.preprocessing import OneHotEncoder
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
overtime_encode = OneHotEncoder(sparse_output = False, drop = 'first', handle_unknown= 'ignore')
overtime_encode.fit(np.array(X_train['OverTime']).reshape(-1,1))

X_train["Overtime_Encoded"]= overtime_encode.transform(np.array(X_train['OverTime']).reshape(-1,1))
X_test["Overtime_Encoded"]= overtime_encode.transform(np.array(X_test['OverTime']).reshape(-1,1))

X_train = X_train.drop(columns='OverTime')
X_test = X_test.drop(columns='OverTime')
```

```python
# Create a StandardScaler
scaler = StandardScaler()


# Fit the StandardScaler to the training data
scaler.fit(X_train)


# Scale the training and testing data
scaled_Xtrain = scaler.transform(X_train)
scaled_Xtest = scaler.transform(X_test)
```

```python
# read y_train data
y_train.info()
```

```python
# Create a OneHotEncoder for the Department column
dept_encode = OneHotEncoder(sparse_output= False, drop= None, handle_unknown = 'ignore')


# Fit the encoder to the training data
dept_encode.fit(np.array(y_train['Department']).reshape(-1,1))


# Create two new variables by applying the encoder
# to the training and testing data
y_dept_train_encode = dept_encode.transform(np.array(y_train['Department']).reshape(-1,1))
y_dept_test_encode = dept_encode.transform(np.array(y_test['Department']).reshape(-1,1))
```

```python
# Create a OneHotEncoder for the Attrition column
encoded_attr = OneHotEncoder(sparse_output= False, drop = 'first', handle_unknown= 'ignore' )

# Fit training data
encoded_attr.fit(np.array(y_train['Attrition']).reshape(-1,1))





# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_attr = encoded_attr.transform(np.array(y_train['Attrition']).reshape(-1, 1))
y_test_attr_encode = encoded_attr.transform(np.array(y_test['Attrition']).reshape(-1,1))

# Display y_train_encoded_attr
y_train_encoded_attr[:5]






```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data and print.
num_of_columns = scaled_Xtrain.shape[1]

print(f'The number of columns in the X training:{num_of_columns}')
```

```python

#Create the input layer
input_layer = layers.Input(shape=(num_of_columns,))


# Create at least two shared layers
shared1_layer = layers.Dense(64, activation = 'relu')(input_layer)
shared2_layer = layers.Dense(32, activation= "relu")(shared1_layer)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_hidden_layer = layers.Dense(16, activation = "relu")(shared2_layer)


# Create the output layer
output_dept = layers.Dense(y_dept_train_encode.shape[1], activation= 'softmax', name = "output_dept")(dept_hidden_layer)

```

```python
y_df
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attr_hidden_layer = layers.Dense(16, activation='relu')(shared2_layer)


# Create the output layer
output_attr = layers.Dense(y_train_encoded_attr.shape[1],activation = 'sigmoid', name = 'output_attr')(attr_hidden_layer)

```

```python
# Create the model
model = Model(inputs = input_layer, outputs= [output_attr, output_dept])

# Compile the model
model.compile(optimizer = 'adam',
              loss ={'output_dept': 'categorical_crossentropy',
                    'output_attr': 'binary_crossentropy'}
              , metrics ={'output_dept': 'accuracy',
              'output_attr': 'accuracy'})

# Summarize the model
model.summary()
```

```python
from re import VERBOSE
# Train the model
history = model.fit(scaled_Xtrain,
                [y_train_encoded_attr, y_dept_train_encode],
                epochs=100,
                batch_size=32,
                validation_split =.2,
                verbose= 2

                )


```

```python
# Evaluate the model with the testing data
evaluation = model.evaluate(scaled_Xtest,
                            [y_dept_test_encode, y_test_attr_encode], return_dict =True)
```
